```python
# mengimport libraries yg diperlukan

import pandas as pd
import numpy as np
```

```python
# mengimport dataset dan mengubah dalam bentuk dataframe (df)

df = pd.read_csv ('dataset2.csv', sep=';')
df
```

```python
# melihat tipe data tiap kolom
print(df.dtypes)
```

```python
# karena manipulasi nilai hanya bisa dilakukan di tipe data string,
# maka ubah semua tipe data kolom menjadi strin
# (karena di sini ingin mengecek panjang nilai tiap baris, untuk deteksi mana yg NIM mana yg bukan)

df['NIM'] = df['NIM'].astype(str)
df['TUGAS1'] = df['TUGAS1'].astype(str)
df['TUGAS2'] = df['TUGAS2'].astype(str)
df['QUIZ1'] = df['QUIZ1'].astype(str)
df['QUIZ2'] = df['QUIZ2'].astype(str)
df['UTS'] = df['UTS'].astype(str)
df['UAS'] = df['UAS'].astype(str)

print(df.dtypes)
```

```python
# jika kolom NIM dan nilai tertukar, maka tukarlah terlebih dahulu nilainya

# cek apakah panjang isi kolom NIM != 10. karena jika bukan 10, maka bukan NIM dan harus ditukar.
mask = (df.NIM.str.len() != 10)

# function untuk tukar nilai NIM dgn kolom lain
def swap():
    df.loc[mask, ['NIM', 'TUGAS1']] = (df.loc[mask, ['TUGAS1', 'NIM']].values)
    df.loc[mask, ['NIM', 'TUGAS2']] = (df.loc[mask, ['TUGAS2', 'NIM']].values)
    df.loc[mask, ['NIM', 'QUIZ1']] = (df.loc[mask, ['QUIZ1', 'NIM']].values)
    df.loc[mask, ['NIM', 'QUIZ2']] = (df.loc[mask, ['QUIZ2', 'NIM']].values)
    df.loc[mask, ['NIM', 'UTS']] = (df.loc[mask, ['UTS', 'NIM']].values)
    df.loc[mask, ['NIM', 'UAS']] = (df.loc[mask, ['UAS', 'NIM']].values)

# panggil function tersebut
# (run cell ini 6 kali, karena ada 6 kolom yang perlu dicek dan ditukar)
swap()
df
```

```python
# jika ada data yang double, maka hapuslah salah satunya dan rapikan kembali tabelnya

# hapus baris yang mengandung NIM duplikat
df = df.drop_duplicates(subset='NIM', keep="first")
df
```

```python
# jika ada tipe data yang tidak sesuai, maka sesuaikan datanya

# samakan semua tipe data menjadi integer
df['NIM'] = pd.to_numeric(df['NIM'])
df['TUGAS1'] = pd.to_numeric(df['TUGAS1'])
df['TUGAS2'] = pd.to_numeric(df['TUGAS2'])
df['QUIZ1'] = pd.to_numeric(df['QUIZ1'])
df['QUIZ2'] = pd.to_numeric(df['QUIZ2'])
df['UTS'] = pd.to_numeric(df['UTS'])
df['UAS'] = pd.to_numeric(df['UAS'])

print(df.dtypes)
```
