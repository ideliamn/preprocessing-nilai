```python
# mengimport libraries yg diperlukan

import pandas as pd
import numpy as np
import csv
```

```python
# mengimport dataset hasil preprocessing dan mengubah dalam bentuk list

with open('dataset2_fixed.csv', newline='') as f:
    reader = csv.reader(f)
    data = list(reader)

print(data)
```

```python
# data sudah berbentuk list

print(type(data))
```

```python
# ubah list menjadi dataframe (df)

df = pd.DataFrame(data[1:],columns=data[0])
df.head()
```

```python
# cek tipe data tiap kolom
# diketahui semua kolom bertipe string (di pandas disebut sbg object)

print(df.dtypes)
```

```python
# karena diketahui tipe data belum integer,
# maka ubah menjadi integer agar bisa melakukan proses statistik

# samakan semua tipe data menjadi integer
df['NIM'] = df['NIM'].astype(int)
df['TUGAS1'] = df['TUGAS1'].astype(int)
df['TUGAS2'] = df['TUGAS2'].astype(int)
df['QUIZ1'] = df['QUIZ1'].astype(int)
df['QUIZ2'] = df['QUIZ2'].astype(int)
df['UTS'] = df['UTS'].astype(int)
df['UAS'] = df['UAS'].astype(int)

print(df.dtypes)
```

```python
# Melakukan perhitungan statistik: max

print('Nilai Maximum')

# mencari berapa jumlah kolom pada df (dataframe)
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range (1, df_len):
    # cari nilai max pada semua kolom
    max_value = df[df.columns[i]].max()
    # cetak hasilnya
    print(df.columns[i],': ',max_value)
    
```

```python
# Melakukan perhitungan statistik: min

print('Nilai Minimum')

# mencari berapa jumlah kolom pada df (dataframe)
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range (1, df_len):
    # cari nilai min pada semua kolom
    min_value = df[df.columns[i]].min()
    # cetak hasilnya
    print(df.columns[i],': ',min_value)
    
```

```python
# Melakukan perhitungan statistik: mean

print('Nilai Mean')

# mencari berapa jumlah kolom pada df (dataframe)
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range (1, df_len):
    # cari nilai mean pada semua kolom,
    # bulatkan (round) menjadi 2 angka di belakang koma
    # lalu cetak hasilnya
    print(df.columns[i],': ',round(df[df.columns[i]].mean(),2))
    
```

```python
# Melakukan perhitungan statistik: var

print('Nilai Varians')

# mencari berapa jumlah kolom pada df (dataframe)
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range (1, df_len):
    # cari nilai var pada semua kolom,
    # bulatkan (round) menjadi 2 angka di belakang koma
    # lalu cetak hasilnya
    print(df.columns[i],': ',round(df[df.columns[i]].var(),2))
    
```

```python
# Melakukan perhitungan statistik: std dev

print('Nilai Standar Deviasi')

# mencari berapa jumlah kolom pada df (dataframe)
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range (1, df_len):
    # cari nilai std pada semua kolom,
    # bulatkan (round) menjadi 2 angka di belakang koma
    # lalu cetak hasilnya
    print(df.columns[i],': ',round(df[df.columns[i]].std(),2))
    
```

```python
# Melakukan perhitungan statistik: median

print('Nilai Median')

# mencari berapa jumlah kolom pada df (dataframe)
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range (1, df_len):
    # cari nilai median pada semua kolom,
    # bulatkan (round) menjadi 2 angka di belakang koma
    # lalu cetak hasilnya
    print(df.columns[i],': ',round(df[df.columns[i]].median(),2))
    
```

```python
# Melakukan perhitungan statistik: modus

print('Nilai Modus')

# mencari berapa jumlah kolom pada df (dataframe)
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range (1, df_len):
    # cari nilai modus pada semua kolom
    # lalu cetak hasilnya
    print(df.columns[i],': ',df.mode()[df.columns[i]][0])
    
```

```python
# Melakukan perhitungan statistik: pencilan atas dan bawah

# import library yg diperlukan
import numpy as np

# mencari berapa jumlah kolom pada df (dataframe)
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range(1, df_len):
    print("Mencari pencilan untuk kolom ",df.columns[i])

    # Hitung nilai Q1 (kuartil 1) dan Q3 (kuartil 3)
    Q1 = np.quantile(df[df.columns[i]], .25)
    Q3 = np.quantile(df[df.columns[i]], .75)

    # Hitung nilai IQR
    IQR = Q3 - Q1

    print('Kuartil 1 = ', Q1)
    print('Kuartil 3 = ', Q3)
    print('IQR = ', IQR)

    min_IQR = Q1 - 1.5 * IQR
    max_IQR = Q3 + 1.5 * IQR

    print('IQR minimum = ', min_IQR)
    print('IQR maksimum = ', max_IQR)

    nilai_min = np.min(df[df.columns[i]])
    nilai_max = np.max(df[df.columns[i]])

    print('Nilai minimum data = ', nilai_min)
    print('Nilai maksimum data = ', nilai_max)

    if (nilai_min < min_IQR):
        print('Ditemukan pencilan bawah')
    else:
        print('Tidak ditemukan pencilan bawah')  

    if (nilai_max > max_IQR):
        print('Ditemukan pencilan atas')
    else:
        print('Tidak ditemukan pencilan atas')

    # buat list untuk menampung pencilan atas (high_out) dan pencilan bawah (low_out)
    low_out = []
    high_out = []

    # simpan ke dalam list tersebut jika ditemukan pencilan atas atau bawah
    for j in df[df.columns[i]]:
        if (j < min_IQR):
            low_out.append(j)
        if (j > max_IQR):
            high_out.append(j)

    # cetak pencilan atas dan bawah jika ditemukan
    print('Pencilan bawah : ', low_out)
    print('Pencilan atas : ', high_out)

    print('\n=============================\n')
```

```python
# menyimpan nilai max, min, mean, dan median tiap kolom ke baris baru di baris paling bawah

# buat dictionary baru untuk menampung data baris baru yg akan ditambahkan
row_max = {}

# kolom NIM diisi dengan nama data perhitungan statistik yg dilakukan
# (pada kode ini adalah mencari nilai max)
row_max['NIM'] = 'MAX'

# mencari berapa jumlah kolom pada df
df_len = len(df.columns)

# lakukan pengulangan pada semua kolom
for i in range (1, df_len):
    # lakukan perhitungan statistik
    # (pada kode ini adalah mencari nilai max)
    max_index = df[df.columns[i]].max()
    # tambahkan hasil perhitungan ke dictionary yg sudah dibuat di atas
    row_max[df.columns[i]] = max_index

#cek apakah isi list sudah berisi data hasil perhitungan
print(row_max)

# semua operasi lainnya seperti min, mean, median menggunakan cara yg sama seperti di atas

row_min = {}
row_min['NIM'] = 'MIN'
df_len = len(df.columns)
for i in range (1, df_len):
    min_index = df[df.columns[i]].min()
    row_min[df.columns[i]] = min_index
print(row_min)

row_mean = {}
row_mean['NIM'] = 'MEAN'
df_len = len(df.columns)
for i in range (1, df_len):
    row_mean[df.columns[i]] = round(df[df.columns[i]].mean(),2)
print(row_mean)

row_median = {}
row_median['NIM'] = 'MEDIAN'
df_len = len(df.columns)
for i in range (1, df_len):
    row_median[df.columns[i]] = round(df[df.columns[i]].median(),2)
print(row_median)

# kumpulkan semua dictionary hasil perhitungan ke dalam satu dictionary baru
rows = [row_max, row_min, row_mean, row_median]

# tambahkan dictionary rows ke df
df = df.append(rows, ignore_index=True)

# lihat df yg sudah ditambahkan datanya
df
```

```python
# Membuat satu kolom baru yang mengakumulasi semua nilai mahasiswa
# (15% * rata-rata tugas + 15% * rata-rata quiz + 35% * UTS + 35% * UAS)

# membuat list kosong yg akan menampung nilai akumulasi
akumulasi = []

# mencari berapa jumlah kolom pada df
df_len = len(df.index)

# lakukan pengulangan pada semua kolom
for i in range (0, df_len):
    # masukkan rumusnya
    rata2_tugas = ((df.at[i,'TUGAS1'])+(df.at[i,'TUGAS2']))/2
    rata2_quiz = ((df.at[i,'QUIZ1'])+(df.at[i,'QUIZ2']))/2
    # tampung dalam variabel total
    total = (0.15*rata2_tugas) + (0.15*rata2_quiz) + (0.35*df.at[i,'UTS']) + (0.35*df.at[i,'UAS'])
    # masukkan nilai total ke list akumulasi
    akumulasi.append(total)
    
# buat kolom baru bernama AKUMULASI, kemudian tambahkan list akumulasi ke dalamnya
df['AKUMULASI'] = akumulasi

# cek df yg sudah ditambahkan datanya
df
```

```python
# Mengekspor dataset hasil proses statistik ke file csv baru
# bernama result_statistik.csv

df.to_csv('result_statistik.csv')
```
