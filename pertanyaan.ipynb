```python
# mengimport libraries yg diperlukan

import pandas as pd
import numpy as np
import csv
```

```python
# mengimport dataset hasil statistik
# kemudian mengubah dalam bentuk dataframe (df)

df = pd.read_csv ('result_statistik.csv', index_col=0)
df.head()
```

# Pertanyaan

```python
# 1. NIM mana yang mendapatkan UTS tertinggi?

top_uts = df['UTS'].max()
print("Nilai tertinggi UTS:",top_uts)
df[df.UTS == top_uts]
```

```python
# 2. NIM mana yang mendapatkan nilai UAS terendah?

low_uas = df['UAS'].min()
print("Nilai terendah UAS:",low_uas)
df[df.UAS == low_uas]
```

```python
# 3. NIM mana sajakah yang mendapatkan nilai quiz-1 di atas rata-rata?

mean_quiz1 = df['QUIZ1'].mean()
print("Mean QUIZ1:",mean_quiz1)

df[df.QUIZ1 > mean_quiz1]
```

```python
# 4. NIM mana sajakah yang mendapatkan nilai quiz-2 di bawah median?

median_quiz2 = df['QUIZ2'].median()
print("Median QUIZ2:",median_quiz2)

df[df.QUIZ2 < median_quiz2]
```

```python
# 5. NIM mana sajakah yang mendapatkan nilai tugas-1 di bawah median?

median_tugas1 = df['TUGAS1'].median()
print("Median TUGAS1:",median_tugas1)

df[df.TUGAS1 < median_tugas1]
```

```python
# 6. NIM mana sajakah yang mendapatkan nilai tugas-2 di atas rata-rata?

median_tugas2 = df['TUGAS2'].mean()
print("Median TUGAS2:",median_tugas2)

df[df.TUGAS2 > median_tugas2]
```

```python
# 8. NIM mana sajakah yang mendapatkan nilai akhir kurang dari mean?

mean_akumulasi = df['AKUMULASI'].mean()
print("Rata-Rata Akumulasi:",mean_akumulasi)

df[df.AKUMULASI < mean_akumulasi]
```
